```python
#we import pandas library to manipulate data
import pandas as pd
```

```python
#we load datasets and print the head to see how they look like
cash_request_df = pd.read_csv('extract - cash request - data analyst.csv')
fees_df = pd.read_csv('extract - fees - data analyst - .csv')

display(cash_request_df.head())
display(fees_df.head())
```

```python
#We print the shape of the CR dataset and check if the number of null values per column makes sense. 
print(f"Cash request df {cash_request_df.shape}")
display(cash_request_df.isnull().sum())

#We check that there are 23970 rows and that there are no CR ids with null values, no missing amount, and creation date columns, reimbursement date and transfer type so it is fine. 
# The other columns have null values but it could make sense depending on other column values. Like for example, user_id columns could be missing because they are deleted_account_id, so
#2103 (null values of user_id) + 21866 (null values of deleted_account_id) = 23969, so looks good. then others like recovery status could have null values if the cash request never had 
#a payment incident. So we should have 20640 cash request with no payment incident. 
```

```python
#We print the shape of the fees dataset and check if the number of null values per column makes sense.
print(f"Fees df {fees_df.shape}")
display(fees_df.isnull().sum())

#here we can see that there are 4 fees not associated to any cash_request id so it needs to be investigated. All other null values could be reasonable as it depends on other columns
```

<span style="font-size:22px;">Getting better understanding of fees dataframe</span>

```python
#to understand better, let's print fees with no userid

#let's recap how it looks like and its null values 
display(fees_df.head())
print(f"Fees df {fees_df.shape}")
display(fees_df.isnull().sum())
```

```python
#fees are in most of the cases associated to a cash request, but there are 4 of them that are not. So let's check those 4.
fees_with_no_CR = fees_df[fees_df['cash_request_id'].isnull()]
display(fees_with_no_CR.head())

#ok looks like those must be fixed as the cash requests appears in the reason. so must be ammended for future analysis of whole datasets
```

```python
#We will replace the cell value of cash_request_id with the number found in the reason column as we saw above. We create a function to it and we will apply this function to all column where 
#null values are found

import re 

def extract_CR_as_float (reason):
    #Search for all number in the string
    match = re.findall(r'\d+', reason)
    # Devuelve el Ãºltimo nÃºmero como float, si existe
    return float(match[-1]) if match else None


#with loc you can choose the index and the column name [i, 'column name']
fees_df.loc[fees_df['cash_request_id'].isna(), 'cash_request_id'] = fees_df['reason'].apply(extract_CR_as_float)

display(fees_df.isnull().sum())

#now we can see that no null values are found in cash_request_id, so it will better later when merging both dataframes
```

```python
fees_df.loc[1911, 'cash_request_id']

#example to check if the function was correctly applied (index 1911 was a null value)
```

```python
#let's check if we have some duplications on id and cash_request_id
print(fees_df.shape)
print(fees_df['id'].nunique())
print(fees_df['cash_request_id'].nunique())

#we have some fees related to the same cash_request id
```

```python
#let's group the dataset by cash request id we count if there are different fees associatd to a cash request 

grouped_df_fees = fees_df.groupby('cash_request_id').agg(id_count=('id', 'count')).reset_index()
display(grouped_df_fees)

#looks like there are cash request with multiple fees
```
